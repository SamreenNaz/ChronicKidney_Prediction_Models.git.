```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/kidney_disease_train.csv")

# Select only numeric columns (excluding 'id' and 'classification')
numeric_df = df.select_dtypes(include=['number'])

# Drop rows with missing values for clean plotting
clean_df = numeric_df.dropna()

# Plot boxplot
plt.figure(figsize=(16, 8))
sns.boxplot(data=clean_df)

# Title and axis formatting
plt.xticks(rotation=45, fontsize=24)  # <-- increased font size here
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
