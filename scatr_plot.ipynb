```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/400 patient data1.csv")

# Drop rows with missing values
df = df.dropna(subset=['bgr', 'bp', 'classification'])

# Separate CKD and NON-CKD for labeling
ckd = df[df['classification'] == 'ckd']
non_ckd = df[df['classification'] == 'notckd']

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(ckd['bgr'], ckd['bp'], color='red', label='CKD')
plt.scatter(non_ckd['bgr'], non_ckd['bp'], color='green', label='NON-CKD')

plt.xlabel('Blood Glucose Random (bgr)')
plt.ylabel('Blood Pressure (bp)')
plt.legend()
plt.show()
```
